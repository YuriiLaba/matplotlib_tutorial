# 2. Basic visualization use case

```python
seattle_weather = pd.read_csv("data/seattle_weather.csv")
austin_weather = pd.read_csv("data/austin_weather.csv")
```

```python
seattle_weather.head()
```

```python
grouped_austin_weather = austin_weather.groupby("DATE").mean().reset_index()
grouped_austin_weather['MONTH'] = grouped_austin_weather['DATE'].apply(lambda x: calendar.month_abbr[x])
```

```python
grouped_seattle_weather = seattle_weather.groupby("DATE").mean().reset_index()
grouped_seattle_weather['MONTH'] = grouped_seattle_weather['DATE'].apply(lambda x: calendar.month_abbr[x])
```

```python
grouped_seattle_weather
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(grouped_seattle_weather["MONTH"], grouped_seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(grouped_austin_weather["MONTH"], grouped_austin_weather["MLY-PRCP-NORMAL"])

plt.show()
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(1, 1, 1)
ax.plot(grouped_seattle_weather["MONTH"], grouped_seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(grouped_austin_weather["MONTH"], grouped_austin_weather["MLY-PRCP-NORMAL"])
plt.show()
```

## Customizing data appearance: color, marker, linestyle

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(grouped_seattle_weather["MONTH"], grouped_seattle_weather["MLY-PRCP-NORMAL"], color='g', marker='o', linestyle='dotted')
ax.plot(grouped_austin_weather["MONTH"], grouped_austin_weather["MLY-PRCP-NORMAL"], color='r', marker='v', linestyle='--')

ax.set_xlabel("Time (months)", c='r', size=30)
ax.set_ylabel("Precipitation (inches)", c='g', size=15)
ax.set_title("Weather patterns in Austin and Seattle", c="c", size=50)

plt.show()
```

## What can we do to see more?

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 10), sharey=True) # do not forget to mention sharey

ax[0].plot(grouped_seattle_weather["MONTH"], grouped_seattle_weather["MLY-PRCP-NORMAL"], color = 'g')
ax[0].plot(grouped_seattle_weather["MONTH"], grouped_seattle_weather["MLY-PRCP-25PCTL"], color = 'g', linestyle = '--')
ax[0].plot(grouped_seattle_weather["MONTH"], grouped_seattle_weather["MLY-PRCP-75PCTL"], color = 'g', linestyle = '--')

ax[1].plot(grouped_austin_weather["MONTH"], grouped_austin_weather["MLY-PRCP-NORMAL"], color = 'r')
ax[1].plot(grouped_austin_weather["MONTH"], grouped_austin_weather["MLY-PRCP-25PCTL"], color = 'r', linestyle = '--')
ax[1].plot(grouped_austin_weather["MONTH"], grouped_austin_weather["MLY-PRCP-75PCTL"], color = 'r', linestyle = '--')

plt.show()
```

# 3. Time series visualization

## 3.1 Plot time series data

```python
climate_change = pd.read_csv("data/climate_change.csv", parse_dates=["date"], index_col="date")
```

```python
climate_change.head()
# co2 - carbon dioxide
# temperature measured per day
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(climate_change.index, climate_change["relative_temp"])
ax.set_xlabel("Time")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

seventies = climate_change["1970-01-01":"1979-12-31"]
ax.plot(seventies.index, seventies["co2"])
plt.show()
```

## 3.2 Plot two time series on the same plot

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(climate_change.index, climate_change["relative_temp"])
ax.plot(climate_change.index, climate_change["co2"])
ax.set_xlabel("Time")
ax.set_ylabel("Relative temperature (Celsius)")
plt.show()
```

## Discover twinx() method

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.set_xlabel("Time")
ax.set_ylabel("co2", color="g")
ax.tick_params('y', colors="g")
ax.plot(climate_change.index, climate_change["co2"], color="g")

ax2 = ax.twinx()

ax2.set_ylabel("relative_temp", color="r")
ax2.tick_params('y', colors="r")
ax2.plot(climate_change.index, climate_change["relative_temp"], color="r")

plt.show()
```

## Don't repeat yourself

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
       axes.plot(x, y, color=color)
       axes.set_xlabel(xlabel)
       axes.set_ylabel(ylabel, color=color)
       axes.tick_params('y', colors=color)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

plot_timeseries(ax, climate_change.index, climate_change["co2"], "green", "Time (years)", "CO2 levels")

ax2 = ax.twinx()

plot_timeseries(ax2, climate_change.index, climate_change["relative_temp"], "red", "Time (years)", "Relative temperature (Celsius)")

plt.show()
```

## 3.3 Annotations

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(climate_change.index, climate_change["relative_temp"])
ax.annotate(">1 degree", (pd.Timestamp('2015-10-06'), 1))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(climate_change.index, climate_change["relative_temp"])

ax.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={"arrowstyle":"->", "color":"green", "linewidth":3})

plt.show()
```

## 4.1 Quantitative comparisons: bar-charts

```python
medals_by_country_2016 = pd.read_csv("data/medals_by_country_2016.csv", index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.bar(medals_by_country_2016.index, medals_by_country_2016["Gold"], label="Gold")
ax.bar(medals_by_country_2016.index, medals_by_country_2016["Silver"], bottom=medals_by_country_2016["Gold"], label="Silver")
ax.bar(medals_by_country_2016.index, medals_by_country_2016["Bronze"], bottom=medals_by_country_2016["Gold"] + medals_by_country_2016["Silver"], label="Bronze")

ax.legend()
plt.show()
```

```python
summer2016
```

```python
mens_rowing = summer2016[summer2016["Sport"] == "Rowing"]
mens_gymnastics = summer2016[summer2016["Sport"] == "Gymnastics"]
```

## Discover histtype and bins parameters

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.hist(mens_rowing["Weight"], label="Rowing", histtype="step", bins=5)

ax.hist(mens_gymnastics["Weight"], label="Gymnastics", histtype="step", bins=5)

ax.set_xlabel("Weight (kg)")
ax.set_ylabel("# of observations")

ax.legend()
plt.show()
```

## 4.4 Statistical plotting

## Add standard deviation info to bar chart

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.bar("Rowing", mens_rowing["Height"].mean(), yerr=mens_rowing["Height"].std())
ax.bar("Gymnastics", mens_gymnastics["Height"].mean(), yerr=mens_gymnastics["Height"].std())
ax.set_ylabel("Height (cm)")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.errorbar(grouped_seattle_weather["MONTH"], grouped_seattle_weather["MLY-TAVG-NORMAL"], yerr=grouped_seattle_weather["MLY-TAVG-STDDEV"])
ax.errorbar(grouped_austin_weather["MONTH"], grouped_austin_weather["MLY-TAVG-NORMAL"], yerr=grouped_austin_weather["MLY-TAVG-STDDEV"])

ax.set_ylabel("Temperature (Fahrenheit)")

plt.show()
```
